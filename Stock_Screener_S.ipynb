```python
!pip install requests beautifulsoup4 pandas
!pip install yfinance pandas

import yfinance as yf
import pandas as pd

# Lista expandida de símbolos de ações para análise
symbols = [
    'AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA', 'FB', 'NVDA', 'NFLX', 'BABA', 'V',
    'JPM', 'JNJ', 'WMT', 'PG', 'MA', 'DIS', 'PYPL', 'UNH', 'HD', 'VZ', 'ADBE',
    'CMCSA', 'NFLX', 'KO', 'PEP', 'PFE', 'MRK', 'T', 'INTC', 'CSCO', 'ORCL',
    'IBM', 'CRM', 'ACN', 'AVGO', 'TXN', 'QCOM', 'AMD', 'MU', 'AMAT', 'LRCX',
    'TSM', 'SPGI', 'MS', 'GS', 'BLK', 'BK', 'SCHW', 'NKE', 'MDT', 'ISRG',
    'MMM', 'GE', 'HON', 'BA', 'LMT', 'NOC', 'GD', 'RTN', 'HUM', 'UNP', 'FDX',
    'UPS', 'CAT', 'DE', 'GM', 'F', 'TM', 'HMC', 'NSANY', 'VOW3.DE', 'BMW.DE',
    'DAI.DE', 'MC.PA', 'SAN.PA', 'BN.PA', 'AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA'
]

# Armazenar dados de ações
stock_data = {}
fundamental_data = {}

# Função para obter dados históricos e fundamentais de ações
def get_stock_and_fundamental_data(symbol):
    stock = yf.Ticker(symbol)
    hist = stock.history(period="1y")
    fundamentals = stock.info
    return hist, fundamentals

for symbol in symbols:
    try:
        hist, fundamentals = get_stock_and_fundamental_data(symbol)
        stock_data[symbol] = hist
        fundamental_data[symbol] = fundamentals
    except Exception as e:
        print(f"Erro ao obter dados para {symbol}: {e}")

# Função para analisar dados e determinar ações para comprar ou vender
def analyze_stocks(stock_data, fundamental_data):
    buy_candidates = []
    sell_candidates = []

    for symbol in symbols:
        if symbol not in stock_data or symbol not in fundamental_data:
            continue

        stock = stock_data[symbol]
        fundamentals = fundamental_data[symbol]

        # Print fundamental data for debugging
        print(f"Analyzing {symbol}:")
        print(f"  P/E Ratio: {fundamentals.get('forwardPE', 'N/A')}")
        print(f"  PEG Ratio: {fundamentals.get('pegRatio', 'N/A')}")
        print(f"  EPS: {fundamentals.get('trailingEps', 'N/A')}")
        print(f"  Profit Margin: {fundamentals.get('profitMargins', 'N/A')}")

        # Critérios para compra
        try:
            if (fundamentals.get('forwardPE', float('inf')) < 20 and
                fundamentals.get('pegRatio', float('inf')) < 2 and
                fundamentals.get('trailingEps', -float('inf')) > 0 and
                fundamentals.get('profitMargins', -float('inf')) > 0.05):
                buy_candidates.append(symbol)
        except ValueError:
            continue

        # Critérios para venda
        try:
            if (fundamentals.get('forwardPE', -float('inf')) > 30 and
                fundamentals.get('pegRatio', -float('inf')) > 2 and
                fundamentals.get('trailingEps', float('inf')) < 0 and
                fundamentals.get('profitMargins', float('inf')) < 0.05):
                sell_candidates.append(symbol)
        except ValueError:
            continue

    return buy_candidates, sell_candidates

# Analisar ações
buy_candidates, sell_candidates = analyze_stocks(stock_data, fundamental_data)

# Exibir resultados
print("Candidatos Fortes para Compra:")
print(buy_candidates)
print("\nCandidatos Fortes para Venda:")
print(sell_candidates)
```
