```python
import numpy as np
import matplotlib.pyplot as plt
def position(t):
  x = 3 * t**2
  y = (2/3) * x + 10
  return x, y
t_values = np.linspace(0, 2, 100)
x_values, y_values = position(t_values)
plt.plot(x_values, y_values, label="Trajetória")
plt.xlabel("x (m)")
plt.ylabel("y (m)")
plt.title("Trajetória da Partícula no Plano xOy")
plt.legend()
plt.grid()
plt.show()
```
